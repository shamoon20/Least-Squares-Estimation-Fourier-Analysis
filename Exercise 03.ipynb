```python
import numpy as np
import matplotlib.pyplot as plt

# Given data from Exercise 1
x = np.array([-0.5, 0.2, 0.7, 1.3, 2.1])
y = np.array([2.3, 3.5, 3.9, 4.5, 5.3])
```

```python
# Least squares estimate using normal equation
w_hat = np.linalg.inv(X.T @ X) @ X.T @ y

a_hat = w_hat[0]
b_hat = w_hat[1]

print("Estimated slope a:", a_hat)
print("Estimated intercept b:", b_hat)
```

Task 4:
The fitted line obtained from the least squares estimate is plotted together with the original data points to visually verify the quality of the fit.

```python
# Plot data points and fitted line
x_line = np.linspace(min(x), max(x), 100)
y_line = a_hat * x_line + b_hat

plt.figure()
plt.scatter(x, y, label="Data points")
plt.plot(x_line, y_line, label="Least squares fit")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

```
